```python
def Random_value_imputation(feature):
    random_values=df[feature].dropna().sample(df[feature].isnull().sum())
    random_values.index=df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature]=random_values
```

```python
for col in num_col:
    Random_value_imputation(col)
```

```python
df[num_col].isnull().sum()
```

```python
Random_value_imputation(' pus cell')
```

```python
df['pus cell clumps'].mode()[0]
```

```python
def impute_mode(feature):
    mode=df[feature].mode()[0]
    df[feature]=df[feature].fillna(mode)
```

```python
for col in cat_col:
    impute_mode(col)
```

## Selecting the Best Features using Feature Selection.

```python
from sklearn.feature_selection import SelectKBest

from sklearn.feature_selection import chi2
```

```python
ind_col= [col for col in df.columns if col!='class']
dep_col='class'
```

```python
X=df[ind_col]
y=df[dep_col]
```

```python
X
```

```python
y
```

```python
ordered_rank_features=SelectKBest(score_func=chi2,k=20)
ordered_feature=ordered_rank_features.fit(X,y)
```

```python
ordered_feature
```

```python
ordered_feature.scores_
```

```python
datascores=pd.DataFrame(ordered_feature.scores_,columns=['scores'])
```

```python
datascores
```

```python
dfcols=pd.DataFrame(X.columns)
dfcols
```

```python
feature_rank=pd.concat([dfcols,datascores],axis=1)
feature_rank
```

```python
feature_rank.columns=['features','score']
feature_rank
```

```python
feature_rank.nlargest(10,'score')
```

```python
selected_columns=feature_rank.nlargest(10,'score')['features'].values
```
